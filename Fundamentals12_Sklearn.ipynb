## 회귀 실습

```python
# 회귀 실습 예제
import numpy as np
import matplotlib.pyplot as plt
r = np.random.RandomState(10)
x = 10 * r.rand(100) # 0~1까지의 난수 100개 생성후 * 100
y = 2 * x - 3 * r.rand(100)
plt.scatter(x,y)
```

```python
x.shape
```

```python
y.shape
```

```python
# 선형회귀 연습예제
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(x,y)
```

### 행렬이 아니라 발생하는 에러

```python
X = x.reshape(100,1)
X[:20]
```

```python
# 모델의 훈련
model.fit(X,y)
```

```python
# 테스트셋 생성
x_new = np.linspace(-1,11,100) # -1 ~ 11까지 숫자 100개 생성
x_new[:20]
```

```python
# 행렬화
X_new = x_new.reshape(100,1)
X_new[:20]
```

```python
#numpy Tip
X_ = x_new.reshape(-1, 1) # -1 means automatically calculate what number is possible
X_.shape[:20]
```

```python
# 모델 예측
y_new = model.predict(X_new)
y_new[:20]
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(X_new ,y_new)
error
```

```python
plt.scatter(x,y,label='input data')
plt.plot(X_new,y_new,color='red',label='regression line')
```

```python
# check how many data is in
print(data.data.shape, data.target.shape)
```

```python
# train 80%
X_train = data.data[:int(len(data.data)*0.8)]
y_train = data.target[:int(len(data.data)*0.8)]
print(len(X_train),len(y_train))
```

```python
# test 20% split
X_test = data.data[int(len(data.data)*0.8):]
y_test = data.target[int(len(data.data)*0.8):]
print(len(X_test),len(y_test))
```

```python
# model fit, predict
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# model accuracy
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

```python
# using train_test_split api
from sklearn.model_selection import train_test_split

result = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

```python
# How train_test_split function return
print(type(result), len(result))
print(result[0].shape, result[1].shape, result[2].shape, result[3].shape)
```
